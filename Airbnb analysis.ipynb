```python
# importing relevant libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import warnings

%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
# Reading data
calendar = pd.read_csv('./data/calendar.csv')
listings = pd.read_csv('./data/listings.csv')
reviews = pd.read_csv('./data/reviews.csv')
```

# Is there a significant difference in price per zip code? 

First, we want to understand if there is a significant price difference between zip codes. Since many factors can contribute to price differences, we will be adjusting the list price per the number of people the property can accomodate. Although not perfect, this measurement will give us a clearer idea of how prices vary per zip code. We will also be looking at apartments only, which is the property type with the largest amout of listings.

```python
listings['property_type'].value_counts()
```

```python
# Creating a new dataframe with required data. 
prices_per_zip = listings.loc[listings.property_type=='Apartment', ['zipcode', 'price', 'accommodates']].copy()
prices_per_zip = prices_per_zip.dropna()
prices_per_zip.head()
```

```python
# Transforming dataset to calculate the desired metrics
prices_per_zip['zipcode'] = prices_per_zip['zipcode'].str[:5]
prices_per_zip['price'] = prices_per_zip['price'].str.replace('[\$\,]', '').astype(float)
prices_per_zip['price_per_person'] = prices_per_zip['price'] / prices_per_zip['accommodates']

#Aggregating by zip code
prices_per_zip = prices_per_zip.groupby(by='zipcode').mean()

# Plotting the price per person per zip code
prices_per_zip['price_per_person'].plot.bar(figsize=(18,10))
```

# How does the number and quality of reviews affect the occupancy rate?

Next we will look at the average ratings per listing_id and try to figure out if there is a correlation between the  occupancy rate and the quantity or quality of the reviews. 

```python
# Creating dataset with occupancy rates per listing_id
occ_rate = calendar[['listing_id', 'available']].replace({'f':1, 't':0}).groupby(by='listing_id').mean().rename(columns={'available':'occupancy'})
occ_rate.head()
```

```python
# Creating dataset with review ratings and occupancy rates. 
listing_ratings = listings[['id', 'number_of_reviews', 'review_scores_rating']]
listing_ratings = listing_ratings.merge(occ_rate, how='inner', left_on='id', right_on='listing_id')
listing_ratings.head()
```

```python
# Plotting review ratings vs occupancy rate
listing_ratings.plot.scatter(x='review_scores_rating', y='occupancy', figsize=(18,10))
```

```python
# Plotting review count vs occupancy rate
listing_ratings.plot.scatter(x='number_of_reviews', y='occupancy', figsize=(18,10))
```

# What are the main factors that contribute to the occpancy rate? 

We understand that several factors can contribute to how often a propery is rented. In this last exercise, we we will select a number of attributes and try to understand the effect that they have in the occupancy rate. 

```python
# Creating a dataset for the regression
x_attr = ['id', 'host_is_superhost', 'price', 'weekly_price', 'security_deposit', 'room_type', 'bedrooms', 'number_of_reviews', 'review_scores_rating', 'instant_bookable']
reg_df = listings[x_attr]
reg_df = reg_df.merge(occ_rate, how='inner', left_on='id', right_on='listing_id')

# Transforming fields
reg_df['price'] = reg_df['price'].str.replace('[\$\,]', '').astype(float)

reg_df['weekly_price'] = reg_df['weekly_price'].str.replace('[\$\,]', '').astype(float)
reg_df['weekly_discount'] = (reg_df['price'] - reg_df['weekly_price'] / 7) / reg_df['price']
reg_df['weekly_discount'] = reg_df['weekly_discount'].fillna(0.0)

reg_df['deposit_required'] = reg_df['security_deposit'].notna()

reg_df = reg_df.drop(columns=['weekly_price', 'security_deposit'])

reg_df = reg_df.dropna()  # Dropping rows with missing information. 
reg_df.head()
```

```python
# Selecting X and y variables. 
y = reg_df['occupancy']
X = reg_df[['host_is_superhost', 'price', 'weekly_discount', 'deposit_required', 'room_type', 'bedrooms', 'number_of_reviews', 'review_scores_rating', 'instant_bookable']]

# Get dummies
cat_cols = [col for col in X.columns if X[col].dtype == 'object']  # List of categorical variables
X = pd.get_dummies(data=X, columns=cat_cols, drop_first=True)  # Get dummies
X.head()
```

```python
# Split data into training and test data, and initialize a linear model
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=.30, random_state=42)
lm_model = LinearRegression(normalize=True)

# Fitting the model
lm_model.fit(X_train, y_train)
```

```python
#Predict using new model
y_test_preds = lm_model.predict(X_test)
y_train_preds = lm_model.predict(X_train)

#Score using new model
test_score = r2_score(y_test, y_test_preds)
train_score = r2_score(y_train, y_train_preds)

print('Train Score: ', train_score, '; Test Score: ', test_score)
```

```python
def coef_weights(coefficients, X_train):
    '''
    INPUT:
    coefficients - the coefficients of the linear model 
    X_train - the training data, so the column names can be used
    OUTPUT:
    coefs_df - a dataframe holding the coefficient, estimate, and abs(estimate)
    
    Provides a dataframe that can be used to understand the most influential coefficients
    in a linear model by providing the coefficient estimates along with the name of the 
    variable attached to the coefficient.
    '''
    coefs_df = pd.DataFrame()
    coefs_df['est_int'] = X_train.columns
    coefs_df['coefs'] = lm_model.coef_
    coefs_df['abs_coefs'] = np.abs(lm_model.coef_)
    coefs_df = coefs_df.sort_values('abs_coefs', ascending=False)
    return coefs_df

#Use the function
coef_df = coef_weights(lm_model.coef_, X_train)

#A quick look at the top results
coef_df
```
